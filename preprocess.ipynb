```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import MiniBatchKMeans
import matplotlib.pyplot as plt



scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

inertia = []
K = range(2, 12)  # test cluster sizes between 2 and 11

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia (SSE)')
plt.title('Elbow Method for Optimal k')
plt.show()
```

```python

from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

K = range(2, 12)
silhouette = []

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    sil = silhouette_score(X_scaled, labels)
    silhouette.append(sil)
    print(f"k={k}, silhouette score={sil:.4f}")

# Plot silhouette scores
plt.plot(K, silhouette, 'ro-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Scores for KMeans')
plt.show()


```

```python
def recommend_from_cluster(track_name, df, n=5):
    # Get the cluster of the input track
    cluster_id = df.loc[df['track_name'] == track_name, 'cluster'].values[0]
    
    # Get other songs from the same cluster
    cluster_songs = df[df['cluster'] == cluster_id]
    
    # Exclude the input song itself
    recommendations = cluster_songs[cluster_songs['track_name'] != track_name]
    
    # Pick top N recommendations
    return recommendations[['track_name', 'artist', 'danceability', 'energy']].head(n)


```

```python
from sklearn.metrics.pairwise import cosine_similarity

def recommend_similar(track_name, df, features, n=5):
    # Extract features matrix
    X = df[features].values
    
    # Compute similarity
    sim_matrix = cosine_similarity(X)
    
    # Map track index
    idx = df[df['track_name'] == track_name].index[0]
    
    # Get similarity scores
    sim_scores = list(enumerate(sim_matrix[idx]))
    
    # Sort by similarity
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    
    # Pick top N excluding itself
    top_idx = [i for i, score in sim_scores[1:n+1]]
    
    return df.iloc[top_idx][['track_name', 'artist', 'danceability', 'energy']]

```

```python
from sklearn.metrics.pairwise import cosine_similarity

def get_recommendations(df, cluster_labels, track_name=None, label=None, 
                        features=['danceability','energy','valence','tempo'], 
                        mode="similar", n=5):
   
    
    if mode == "similar":
        if track_name is None:
            raise ValueError("Provide a track_name for 'similar' mode.")
        
        # Feature matrix
        X = df[features].values
        sim_matrix = cosine_similarity(X)
        
        idx = df[df['track_name'] == track_name].index[0]
        sim_scores = list(enumerate(sim_matrix[idx]))
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        
        top_idx = [i for i, score in sim_scores[1:n+1]]
        return df.iloc[top_idx][['track_name','artist','cluster']]
    
    elif mode == "cluster":
        if track_name is None:
            raise ValueError("Provide a track_name for 'cluster' mode.")
        
        cluster_id = df.loc[df['track_name'] == track_name, 'cluster'].values[0]
        cluster_songs = df[df['cluster'] == cluster_id]
        
        return cluster_songs[cluster_songs['track_name'] != track_name][['track_name','artist']].head(n)
    
    elif mode == "playlist":
        if label is None:
            raise ValueError("Provide a label for 'playlist' mode.")
        
        matching_clusters = [c for c, l in cluster_labels.items() if l == label]
        return df[df['cluster'].isin(matching_clusters)][['track_name','artist']].sample(n)
    
    else:
        raise ValueError("Mode must be 'similar', 'cluster', or 'playlist'.")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

def get_recommendations(df, cluster_labels, track_idx=None, label=None, 
                        features=['danceability','energy','valence','tempo'], 
                        mode="similar", n=5):
    """
    Recommendation system for your Spotify dataset (memory-efficient).
    """
    
    X = df[features].values  # feature matrix
    
    if mode == "similar":
        if track_idx is None:
            raise ValueError("Provide a track_idx (row number) for 'similar' mode.")
        
        # Compute similarity only for one row vs all
        target = X[track_idx].reshape(1, -1)
        sim_scores = cosine_similarity(target, X)[0]  # shape = (95837,)
        
        # Get top n indices (excluding the track itself)
        top_idx = np.argsort(sim_scores)[::-1][1:n+1]
        
        recs = df.iloc[top_idx][['name_song','name_artists','cluster','Cluster_Label'] + features]
        return recs.reset_index(drop=True)
    
    elif mode == "cluster":
        if track_idx is None:
            raise ValueError("Provide a track_idx for 'cluster' mode.")
        
        cluster_id = df.iloc[track_idx]['cluster']
        cluster_songs = df[df['cluster'] == cluster_id].drop(track_idx)
        recs = cluster_songs.sample(min(n, len(cluster_songs)))[['name_song','name_artists','cluster','Cluster_Label'] + features]
        return recs.reset_index(drop=True)
    
    elif mode == "playlist":
        if label is None:
            raise ValueError("Provide a label for 'playlist' mode (e.g., 'Happy/Pop').")
        
        matching_clusters = [c for c, l in cluster_labels.items() if l == label]
        recs = df[df['cluster'].isin(matching_clusters)].sample(n)[['name_song','name_artists','cluster','Cluster_Label'] + features]
        return recs.reset_index(drop=True)
```
